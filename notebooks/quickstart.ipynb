# qreals quickstart

`qreals` computes q-deformed rational and real numbers via the
Morier-Genoud-Ovsienko (MGO) continued-fraction construction. This notebook
installs the package and runs three worked examples: the stable q-series of a
real number, the arithmetic deficit, and the one-sided jump gap of a rational.

Run the cells top to bottom. The first cell installs `qreals` only if it is not
already present, so the notebook works on Google Colab, on Binder, and locally.

```python
# Install qreals if it is not already available (for example on a fresh Colab
# runtime). Once v0.1.0 is on PyPI this is simply:  pip install qreals
# Until that first release is live, the git install below works today.
import importlib.util
import subprocess
import sys

if importlib.util.find_spec("qreals") is None:
    subprocess.run(
        [
            sys.executable, "-m", "pip", "install", "--quiet",
            "git+https://github.com/patrickt6/qreals.git",
        ],
        check=True,
    )

import qreals

print("qreals version:", qreals.__version__)
```

## 1. The q-series of a real number

`q_real_truncated(x, N)` returns the first `N` power-series coefficients of
`[x]_q` for any real `x` you can write down, with the MGO Proposition 1.1
guarantee that those `N` coefficients are stable: they do not change if you ask
for more. The input is a string parsed by sympy, so `"pi"`, `"sqrt(2)"`,
`"(1+sqrt(5))/2"` (the golden ratio), and `"3/2"` all work.

```python
print("[pi]_q      ", qreals.q_real_truncated("pi", 12))
print("[sqrt(2)]_q ", qreals.q_real_truncated("sqrt(2)", 12))
print("[golden]_q  ", qreals.q_real_truncated("(1+sqrt(5))/2", 12))

# A rational has an exact q-analog, a rational function in q:
print("[3/2]_q     ", qreals.q_rational(3, 2))
```

## 2. The arithmetic deficit

The map `x -> [x]_q` is not a ring homomorphism, so `[x]_q + [y]_q` is not
`[x + y]_q`. `deficit(x, y, op, N)` measures exactly that gap for `op` `"+"` or
`"*"`. For rational inputs it also returns the exact closed form in q, and it
reports the two sanity values `D(1) = 0` (both sides agree at q = 1) and, for a
sum of `x, y >= 1`, `D(0) = -1`.

```python
d = qreals.deficit("3/2", "5/2", "+", 8)
print("deficit coefficients:", d.deficit)
print("exact closed form:   ", d.exact)             # q**3 - 1 for these inputs
print("D(1), D(0):          ", (d.deficit_at_q1, d.deficit_at_q0))
```

## 3. The jump gap of a rational

A rational `p/s` has two q-versions: the right value `[p/s]_q^+` (the limit from
above, equal to `q_rational(p, s)`) and the left value `[p/s]_q^-` (the limit
from below). `jumpgap(p, s)` returns both and the factored gap between them,
following Jouteur (arXiv:2503.02122).

```python
j = qreals.jumpgap(3, 2)
print("right [3/2]_q^+:", j.right)
print("left  [3/2]_q^-:", j.left)
print("gap:           ", j.gap)
```
